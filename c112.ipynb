```python
#imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
#the function we are plotting
#k must be an integer!
def RCS(k,a,b,sigma_squared):
    k = int(k) #just to make sure
    if k==0:
        return sigma_squared / (1-np.abs(a)**2) * (np.abs(1-b*np.conj(a)) + np.abs(b)**2 * (1 - np.abs(a)**2) )
    elif k>=1:
        return sigma_squared / (1-np.abs(a)**2) * ((1-b/a) * (1 - np.conj(b)*a)) * (-a)**k
    else:
        return np.conj(RCS(-k, a, b, sigma_squared))

    
```

```python
def plot_for_values(a,b,sigma_squared):
    infunc = [k for k in range(-20,20)]
    y = []
    for i in range(len(infunc)):
        y.append(RCS(infunc[i], a, b, sigma_squared))

    plt.plot(infunc,y)
    plt.xlabel('k')
    plt.ylabel('r(k)')

    plt.title(f"RCS with $a_1$={a}, $b_1$={b}, and $\sigma^2={sigma_squared}$")

    plt.show()    

```

```python
#We plot what we care about. Focusing on a here

plot_for_values(0.75, 1 , 1)
plot_for_values(0.5, 1 , 1)
plot_for_values(0.25, 1 , 1)
plot_for_values(0.1, 1 , 1)
```
